```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("大人")
    else:
      print("大人でない")

array = []

personal_data1 = Human("A", 22)
personal_data2 = Human("B", 19)
personal_data3 = Human("C", 30)

array.append(personal_data1)
array.append(personal_data2)
array.append(personal_data3)

for array_data in array:
  array_data.check_adult()

```
